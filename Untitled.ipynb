```python
def select_all_links(url : str):
    # Pega todos os links únicos de um site
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    links = set()
    for link in soup.find_all('a', attrs={'href': re.compile("^http.?://")}):
        if not re.search('/P[0-9]{1,4}', link.get('href'),  flags=re.IGNORECASE): # chape case
            links.add(link.get('href'))   
    return links
```

```python
def select_all_links_with_date(url : str):
    """
    AVAI Ok
    """
    # Pega todos os links únicos de um site
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    dic = {}

    for div in soup.find_all('div', attrs={'class': 'gdlr-standard-style'}):
        link = div.find_all('a', attrs={'href': re.compile("^http.?://")})[0].get('href')
        if not(dic.get(link)):
            dic[link] = re.search('[0-9]+.[a-z]{3,}.[0-9]{4}', div.get_text(), flags=re.IGNORECASE)[0]
               
    return dic
```

```python
def get_context(url : str, search_expressions : dict, hist : dict, get_date : bool = False):
    """ 
    url : link do página procuradada 
    search_expressions : dicionário contendo a palavra desejada como chave e a regular expression como valor 
    """
    
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    contents = [soup('title')[0], [], '', []] # 0 : Título,  1 : contexto, 2 : Data, 3: keywords
    
    for par in soup.find_all('p'):
        for name, expr in search_expressions.items():
            keyword = re.findall(expr, par.get_text(), flags=re.IGNORECASE)
            if keyword:
                contents[3].append(name)
                hist[name] += len(keyword) # contents[1] --> keywords
                contents[1].append(par.get_text()) # contents[2 ] --> contexto
    
    if get_date:
        date = None
        date_search = ['(?:janeiro|fevereiro|março|abril|maio|junho|julho|agosto|setembro|outubro|novembro|dezembro)\s[0-9]+,\s[0-9]{4}',
                      '[0-9]{1,2}/[0-9]{1,2}/[0-9]{4}']
        [s.extract() for s in soup(['style', 'script', '[document]', 'head', 'title'])]

        for date_exp in date_search:
            date = re.search(date_exp, soup.getText(), flags=re.IGNORECASE) 
            if date:
                contents[2] = date[0]
                break
        # assert date
        
    return contents, hist
                           
```
